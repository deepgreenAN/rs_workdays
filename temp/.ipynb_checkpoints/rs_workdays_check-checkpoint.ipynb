### extract_workdays_intraday_jp

```python
with open("temp/naive_stock_df.pickle", "rb") as f:
    naive_stock_df = pickle.load(f)
    
x = np.arange(0, len(naive_stock_df))
p = bokeh.plotting.figure(plot_width=1000, plot_height=500)
p.line(x, naive_stock_df["Close_6502"])
show(p)
```

```python
index_value = naive_stock_df.index.values.astype(np.int64)
index_value
```

```python
fixed_index_value = (index_value / 1.e9).astype(np.int64)
fixed_index_value
```

```python
extracted_bools = extract_workdays_intraday_jp_bool_numpy_rs(fixed_index_value)
extracted_bools
```

```python
extracted_stock_df = naive_stock_df.loc[extracted_bools,:]
```

```python
x = np.arange(0, len(extracted_stock_df))
p = bokeh.plotting.figure(plot_width=1000, plot_height=500)
p.line(x, extracted_stock_df["Close_6502"])
show(p)
```

```python
%%timeit 
extracted_bools = extract_workdays_intraday_jp_bool_numpy_rs(fixed_index_value)
```
